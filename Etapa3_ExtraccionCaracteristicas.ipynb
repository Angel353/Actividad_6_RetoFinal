```python
# Importamos las librerías pandas, numpy, matplotlib respectivamente

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from funpymodeling.exploratory import freq_tbl
```

```python
# Carga desde un archivo .csv

df_vistazo = pd.read_csv('Datos_limpiosDF.csv')

# Eliminación de columnas inservibles
df_total = df_vistazo.drop(['Unnamed: 0'], axis=1)
df_total
```

```python
# Elección de variables (10) para este dataframe
df_etapa3 = df_vistazo[['host_response_time', 'host_identity_verified', 'number_of_reviews', 'review_scores_cleanliness', 'review_scores_location', 'host_acceptance_rate', 'property_type', 'availability_365', 'review_scores_communication', 'review_scores_value']]
df_etapa3
```

```python
#Identificar valores nulos por columna
valores_nulos= df_etapa3.isnull().sum()
valores_nulos
```

```python
# Análisis univariado de las variables categóricas (Retailer)

freq_tbl(df_etapa3)
```

```python
# Obtengo un análisis univariado de una variable categórica en específico
table1 = freq_tbl(df_etapa3['host_response_time'])
table1

# Obtengo un filtro de los valores más relevantes de las variables catégoricas seleccionada

Filtro1 = table1[(table1['frequency']> 1) & ~(table1["host_response_time"] == "Desconocido")]
Filtro1

Filnot1 = Filtro1.drop(['percentage','cumulative_perc'], axis=1)
# Ajusto el índice de mi dataframe
Filtro_index1 = Filnot1.set_index('host_response_time')
Filtro_index1

# La frecuencia de desconocidos fue de 3091 (12.7%)
# La respuesta del host que dura un estimado de una hora en responder pertenece al 70.1%
```

```python
# Realizamos un gráfico de barras del dataframe filtrado
Filtro_index1.plot(kind = 'bar', width = 1, figsize = (10,4))
plt.title('Tiempo de respuesta del host')
plt.xlabel('Tiempo de respuesta')
plt.ylabel('Frecuencia')
```
