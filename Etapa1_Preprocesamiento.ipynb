```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_vistazo = pd.read_excel('DF_Mexico.xlsx')
data_vistazo
```

```python
#Identificar valores nulos por columna
valores_nulos=data_vistazo.isnull().sum()
valores_nulos
```

```python
data_vistazo.info()
```

```python
# Se eliminarán estás variables ya que no contienen datos dentro del data frame

data_vistazo_2= data_vistazo.drop(['neighbourhood_group_cleansed', 'bathrooms', 'calendar_updated', 'license'], axis=1)
data_vistazo_2
```

```python
# Elegimos solo los datos cualitativos y los separamos de los datos cuantitativos.
datos_cuali = data_vistazo_2.select_dtypes(include=['object'])
datos_cuali

# Elegimos solo los datos cualitativos y los separamos de los datos cuantitativos.
datos_time = data_vistazo_2.select_dtypes(include=['datetime64[ns]'])
datos_time
```

```python
# Vemos la información del data frame
datos_cuali.info()
datos_time.info()
```

```python
#Identificar valores nulos por (fecha)
valores_nulos_datetime=datos_time.isnull().sum()
valores_nulos_datetime
```

```python
#Identificar valores nulos por (cualitativos)
valores_nulos_cuali=datos_cuali.isnull().sum()
valores_nulos_cuali
```

```python
# Reemplazamos valores nulos de tiempo en el dataframe con "desconocido" puesto que no existe ningun patrón que indique el tiempo exacto
df_tiempo =datos_time.fillna("Desconocido") 
df_tiempo

# Reemplazamos valores nulos cualitativos en el dataframe con "desconocido" puesto que no existe ningun patrón para encontrar los datos exactos

df_cuali_in =datos_cuali.fillna("Desconocido") 
df_cuali_in
```

```python
# Juntamos los data frames cualitativos en uno solo
df_cuali = pd.concat([df_tiempo, df_cuali_in], axis=1)
```

```python
#Identificar valores nulos por (cualitativos)
valores_nulos_1=df_cuali.isnull().sum()
valores_nulos_1
```

```python
# Separamos los datos cuantitativos también
datos_cuanti = data_vistazo_2.select_dtypes(np.number)
datos_cuanti.head()
```

```python
# Vemos la información del data frame
datos_cuanti.info()
```

```python
#Identificar valores nulos por (cuantitativos)
valores_nulos_2=datos_cuanti.isnull().sum()
valores_nulos_2
```

```python
# Copiamos el data frame cuantitativo

datos_cuanti2 = datos_cuanti.copy()
```

```python
#Sustituir valores nulos con promedio 
# Estos valores están en porcentaje y por lo tanto se decide usar promedio para considerar todos los datos dentro de una escala contando ya con un parámetro definido 
datos_cuanti2["host_response_rate"]=datos_cuanti["host_response_rate"].fillna(round(datos_cuanti["host_response_rate"].mean(),1))
datos_cuanti2["host_acceptance_rate"]=datos_cuanti["host_acceptance_rate"].fillna(round(datos_cuanti["host_acceptance_rate"].mean(),1))
# Estos otros valores tienen parámetro del 1 al 5 por lo tanto también se puede usar la media
datos_cuanti2["review_scores_rating"]=datos_cuanti["review_scores_rating"].fillna(round(datos_cuanti["review_scores_rating"].mean(),1))
datos_cuanti2["review_scores_accuracy"]=datos_cuanti["review_scores_accuracy"].fillna(round(datos_cuanti["review_scores_accuracy"].mean(),1))
datos_cuanti2["review_scores_cleanliness"]=datos_cuanti["review_scores_cleanliness"].fillna(round(datos_cuanti["review_scores_cleanliness"].mean(),1))
datos_cuanti2["review_scores_checkin"]=datos_cuanti["review_scores_checkin"].fillna(round(datos_cuanti["review_scores_checkin"].mean(),1))
datos_cuanti2["review_scores_communication"]=datos_cuanti["review_scores_communication"].fillna(round(datos_cuanti["review_scores_communication"].mean(),1))
datos_cuanti2["review_scores_location"]=datos_cuanti["review_scores_location"].fillna(round(datos_cuanti["review_scores_location"].mean(),1))
datos_cuanti2["review_scores_value"]=datos_cuanti["review_scores_value"].fillna(round(datos_cuanti["review_scores_value"].mean(),1))

#Sustituir valores nulos con mediana
# Para esta cáptura de datos no hay parametro y el rellenado de nulos no se verá tan afectado como sí al usar media
datos_cuanti2["bedrooms"]=datos_cuanti["bedrooms"].fillna(round(datos_cuanti["bedrooms"].median(),1))
datos_cuanti2["beds"]=datos_cuanti["beds"].fillna(round(datos_cuanti["beds"].median(),1))
datos_cuanti2["minimum_minimum_nights"]=datos_cuanti["minimum_minimum_nights"].fillna(round(datos_cuanti["minimum_minimum_nights"].median(),1))
datos_cuanti2["maximum_minimum_nights"]=datos_cuanti["maximum_minimum_nights"].fillna(round(datos_cuanti["maximum_minimum_nights"].median(),1))
datos_cuanti2["minimum_maximum_nights"]=datos_cuanti["minimum_maximum_nights"].fillna(round(datos_cuanti["minimum_maximum_nights"].median(),1))
datos_cuanti2["maximum_maximum_nights"]=datos_cuanti["maximum_maximum_nights"].fillna(round(datos_cuanti["maximum_maximum_nights"].median(),1))
datos_cuanti2["minimum_nights_avg_ntm"]=datos_cuanti["minimum_nights_avg_ntm"].fillna(round(datos_cuanti["minimum_nights_avg_ntm"].median(),1))
datos_cuanti2["maximum_nights_avg_ntm"]=datos_cuanti["maximum_nights_avg_ntm"].fillna(round(datos_cuanti["maximum_nights_avg_ntm"].median(),1))
datos_cuanti2["maximum_nights_avg_ntm"]=datos_cuanti["maximum_nights_avg_ntm"].fillna(round(datos_cuanti["maximum_nights_avg_ntm"].median(),1))
datos_cuanti2["reviews_per_month"]=datos_cuanti["reviews_per_month"].fillna(round(datos_cuanti["reviews_per_month"].median(),1))
```

```python
#Identificar valores nulos por (cuantitativos)
valores_nulos_3=datos_cuanti2.isnull().sum()
valores_nulos_3
```

```python
# Copiamos el data frame cuantitativo sin valores nulos

df_cuanti = datos_cuanti2.copy()
```

```python
# Guardamos las medidas que no tienen outliers dado los parámetros en otro data frame

df_medidas = df_cuanti[['latitude', 'longitude','host_response_rate','host_acceptance_rate', 'review_scores_rating','review_scores_accuracy','review_scores_cleanliness', 'review_scores_checkin','review_scores_communication','review_scores_location', 'review_scores_value','price','id','scrape_id','host_id']]
df_medidas

# Se las quitamos a los datos originales limpios

df_cuanti_out = df_cuanti.drop(['latitude', 'longitude','host_response_rate','host_acceptance_rate', 'review_scores_rating','review_scores_accuracy','review_scores_cleanliness', 'review_scores_checkin','review_scores_communication','review_scores_location', 'review_scores_value','price','id','scrape_id','host_id'], axis=1)
```

```python
# Verificamos valores atípicos por plot (diagrama de caja o bigote) de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
df_cuanti_out.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y= df_cuanti_out

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= df_cuanti_out[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios del Dataframe
data3_iqr= df_cuanti_out[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data3_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data4_iqr=data3_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

```python
#Corroboramos valores nulos del dataframe LIMPIO sin outliers
valores_nulos=data4_iqr.isnull().sum()
valores_nulos
```

```python
# Unimos las columnas cualitativas y las que dado su naturaleza no poseían outliers con el dataframe

Datos_no_outliers = pd.concat([df_cuali, df_medidas], axis=1)
Datos_limpios = pd.concat([Datos_no_outliers, data4_iqr], axis=1)
Datos_limpios
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos4= Datos_no_outliers.isnull().sum()
valores_nulos4
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulosFinal= Datos_limpios.isnull().sum()
valores_nulosFinal
```
