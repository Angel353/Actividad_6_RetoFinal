```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carga de archivo
df_vistazo = pd.read_csv('Datos_limpiosDF.csv')
df_vistazo
```

```python
df = df_vistazo.drop(['Unnamed: 0'], axis=1)
df
```

```python
# Cambiar a tipo float para trabajar los datos
df['host_acceptance_rate'] = df['host_acceptance_rate'].astype(float)
df['host_acceptance_rate']
```

```python
# Filtro 1: host_acceptance_rate> 50%

filtro1=df[df["host_acceptance_rate"] > 0.5]
filtro1
```

```python
# Filtro 2: Sólo los registros con categoría “superhost”

filtro2=df[df["host_is_superhost"] == 't']
filtro2
```

```python
# Filtro 3: Sólo los registros que no hallan verificado identidad “identity_verified”

filtro3=df[df["host_identity_verified"] == 'f']
filtro3
```

```python
# Filtro 4: Los registro cuyo property_type = “Private room” y “Hotel room”

filtro4=df[(df["room_type"] == "Hotel room") | (df["room_type"] == "Private room")]
filtro4
```

```python
# Filtro 5: Los registros que cuenten con bathroom > 1

# La variable "bathroom" fue removida de la base de datos dado que no había valores en la columna
```

```python
# Filtro 5.1: Los registros que cuenten con minimum_nights_avg_ntm > 1

filtro5=df[df["minimum_nights_avg_ntm"] > 5.6]
filtro5
```

```python
# Filtro 6: Los registros cuyo precio sea mayor de $10,000 y que sean de tipo “Entire home”

filtro6 =df[(df["price"] > 10000) & (df["property_type"] == "Entire home")]
filtro6
```

```python
# Filtro 7: Los registros cuyo review_scores_cleanliness > 4.5

filtro7 = df[df["review_scores_cleanliness"] > 4.5]
filtro7
```

```python
# Filtro 8: Los registros cuyo review_scores_value > 4.9

filtro8 = df[df["review_scores_value"] > 4.9]
filtro8
```

```python
# Filtro 9: Los registros cuya availability_365< 100

filtro9 = df[df["availability_365"] < 100]
filtro9
```

```python
# Filtro 10: Los registros cuya host_response_time sea “within an hour”

filtro10 = df[df["host_response_time"] == "within an hour"]
filtro10
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro1.csv")
filtro2.to_csv("Filtro2.csv")
filtro3.to_csv("Filtro3.csv")
filtro4.to_csv("Filtro4.csv")
filtro5.to_csv("Filtro5.csv")
filtro6.to_csv("Filtro6.csv")
filtro7.to_csv("Filtro7.csv")
filtro8.to_csv("Filtro8.csv")
filtro9.to_csv("Filtro9.csv")
filtro10.to_csv("Filtro10.csv")
```
